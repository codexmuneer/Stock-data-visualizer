```python
import quandl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

quandl.ApiConfig.api_key = 'k5hpQXTQmkr3f2cBYTDb'
names_of_company = ['HBL','ABL', 'UBL', 'BOP']

results = pd.DataFrame()

for names in names_of_company:
    x = quandl.get('PSX/%s' %names, start_date='2011-1-1',end_date='2022-5-1')
    x['company'] = names
    results = results.append(x)
```

```python
# declared all companies dataframes inside other variables
c1 = results[results['company'] == 'HBL']
c2 = results[results['company'] == 'ABL']
c3 = results[results['company'] == 'UBL']
c4 = results[results['company'] == 'BOP']
```

# Task 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (12,12))

ax1.set_title("   Total turnover of stock being traded of HBL")
ax1.set_xlabel("days", fontsize = 20)
ax1.set_ylabel("turnover", fontsize = 20)
ax1.plot(c1.index,c1['Turnover'], color="orange")

ax2.set_title("   Total turnover of stock being traded of ABL")
ax2.set_xlabel("days", fontsize = 20)
ax2.set_ylabel("turnover", fontsize = 20)
ax2.plot(c2.index,c2['Turnover'], color="green")

ax3.set_title("   Total turnover of stock being traded of UBL")
ax3.set_xlabel("days", fontsize = 20)
ax3.set_ylabel("turnover", fontsize = 20)
ax3.plot(c3.index,c3['Turnover'], color="blue")

ax4.set_title("   Total turnover of stock being traded of BOP")
ax4.set_xlabel("days", fontsize = 20)
ax4.set_ylabel("turnover", fontsize = 20)
ax4.plot(c4.index,c4['Turnover'], color="magenta")
```

# Task 2

```python
# it will calculate daily price change in stock for all companies
h = c1.loc[:,["Open","Close"]].diff()
a = c2.loc[:,["Open","Close"]].diff()
u = c3.loc[:,["Open","Close"]].diff()
b = c4.loc[:,["Open","Close"]].diff()
#plotting all companies graph
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (12,12))

ax1.set_title("   Daily Price change in stock of HBL")
ax1.set_xlabel("Years", fontsize = 20)
ax1.set_ylabel("stock price", fontsize = 20)
ax1.plot(h, color="orange")

ax2.set_title("   Daily Price change in stock of ABL")
ax2.set_xlabel("Years", fontsize = 20)
ax2.set_ylabel("stock price", fontsize = 20)
ax2.plot(a, color="green")

ax3.set_title("   Daily Price change in stock of UBL")
ax3.set_xlabel("Years", fontsize = 20)
ax3.set_ylabel("stock price", fontsize = 20)
ax3.plot(u, color="blue")

ax4.set_title("   Daily Price change in stock of BOP")
ax4.set_xlabel("Years", fontsize = 20)
ax4.set_ylabel("stock price", fontsize = 20)
ax4.plot(b, color="magenta")
```

# Task 3

```python
m1 = c1.groupby(pd.PeriodIndex(c1.index, freq="M"))['Close'].mean().tolist()
y1 = c1.groupby(pd.PeriodIndex(c1.index, freq="Y"))['Close'].mean().tolist()

m2 = c2.groupby(pd.PeriodIndex(c2.index, freq="M"))['Close'].mean().tolist()
y2 = c2.groupby(pd.PeriodIndex(c2.index, freq="Y"))['Close'].mean().tolist()

m3 = c3.groupby(pd.PeriodIndex(c3.index, freq="M"))['Close'].mean().tolist()
y3 = c3.groupby(pd.PeriodIndex(c3.index, freq="Y"))['Close'].mean().tolist()

m4 = c4.groupby(pd.PeriodIndex(c4.index, freq="M"))['Close'].mean().tolist()
y4 = c4.groupby(pd.PeriodIndex(c4.index, freq="Y"))['Close'].mean().tolist()

fig, ((ax1, ax2), (ax3, ax4), (ax5,ax6),(ax7,ax8)) = plt.subplots(4, 2, figsize = (14,14))

ax1.set_title("   Monthly mean of HBL")
ax1.set_xlabel("x axis")
ax1.set_ylabel("y axis")
ax1.plot(m1, color="orange")

ax2.set_title("    Yearly mean  of HBL")
ax2.set_xlabel("x axis")
ax2.set_ylabel("y axis")
ax2.plot(y1, color="orange")

ax3.set_title("    Monthly mean  of ABL")
ax3.set_xlabel("x axis")
ax3.set_ylabel("y axis")
ax3.plot(m2, color="green")

ax4.set_title("    Yearly mean  of ABL")
ax4.set_xlabel("x axis")
ax4.set_ylabel("y axis")
ax4.plot(y2, color="green")


ax5.set_title("    Monthly mean  of UBL")
ax5.set_xlabel("x axis")
ax5.set_ylabel("y axis")
ax5.plot(m3, color="blue")

ax6.set_title("    Yearly mean  of UBL")
ax6.set_xlabel("x axis")
ax6.set_ylabel("y axis")
ax6.plot(y3, color="blue")

ax7.set_title("    Monthly mean  of BOP")
ax7.set_xlabel("x axis")
ax7.set_ylabel("y axis")
ax7.plot(m4, color="magenta")

ax8.set_title("    Yearly mean  of BOP")
ax8.set_xlabel("x axis")
ax8.set_ylabel("y axis")
ax8.plot(y4, color="magenta")
```

# Task 4

```python
drh = c1['Close'].pct_change()
dra = c2['Close'].pct_change()
dru = c3['Close'].pct_change()
drb = c4['Close'].pct_change()

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (12,12))

ax1.set_title("   Daily returns of HBL")
ax1.set_xlabel("Days", fontsize = 20)
ax1.set_ylabel("closing Price", fontsize = 20)
ax1.plot(drh, color="orange")

ax2.set_title("   Daily returns of ABL")
ax2.set_xlabel("Days", fontsize = 20)
ax2.set_ylabel("closing Price", fontsize = 20)
ax2.plot(dra, color="green")

ax3.set_title("   Daily returns of UBL")
ax3.set_xlabel("Days", fontsize = 20)
ax3.set_ylabel("closing Price", fontsize = 20)
ax3.plot(dru, color="blue")

ax4.set_title("   Daily returns of BOP")
ax4.set_xlabel("Days", fontsize = 20)
ax4.set_ylabel("closing Price", fontsize = 20)
ax4.plot(drb, color="magenta")
```
